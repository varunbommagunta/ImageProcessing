## **Sci-Kit Image**

```python
from skimage import io,img_as_float
```

```python
img = io.imread("/content/dog-breeds.png")
```

```python
type(img)
```

```python
plt.imshow(img)
plt.xticks([])
plt.yticks([]);
```

**NOTE:**

```python
img[:,:,0]
```

```python
img_np_float = io.imread("/content/dog-breeds.png").astype(np.float32)
```

```python
img_np_float[:,:,0]
```

```python
img_float = io.imread("/content/dog-breeds.png")
img_float = img_as_float(img) #Normalizes between 0 to 1
```

```python
img_float[:,:,0]
```

## **Reading Multiple Images**

```python
import glob
```

```python
path = '/content/drive/MyDrive/Pets/*'
```

```python
for file in glob.glob(path):
  img = cv2.imread(file)
  cv2_imshow(img)
```
