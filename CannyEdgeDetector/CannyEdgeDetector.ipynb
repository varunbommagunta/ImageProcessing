```
import cv2   #opencv
import numpy as np
from matplotlib import pyplot as plt
import viz  #python file containing visualization function
```

```
img = cv2.imread('Toucan.jpg',0)
```

```
edges = cv2.Canny(img,100,200)  #Canny edge detector,(Internally applies (5,5) Gaussian kernal to first smooth it)
```

```
viz.show(img,edges)
```

```
blur = cv2.GaussianBlur(img,(7,7),1)  #Blur it first then apply canny edge detector || Params :(image,(kernel_size,kernel_size),std.dev)
b_edges = cv2.Canny(blur,42,200)
```

```
viz.show(img,b_edges)
```
