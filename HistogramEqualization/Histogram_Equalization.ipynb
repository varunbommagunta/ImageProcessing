## **Histogram Equalization**

```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('/content/Unequalized_landscape.jpg',0)  #Read the image in gray 
```

```python
img = cv2.resize(img,(400,400),interpolation = cv2.INTER_LINEAR)  #Resize the image 
```

```python
cv2_imshow(img)
```

```python
plt.hist(img.ravel(),bins = 256,range = [0,256])  #Plotting the histogram for number of pixels
plt.xlabel("Pixel Values",size = 16)
plt.ylabel("Number of Pixels",size = 16)
plt.show()
```

```python
equ = cv2.equalizeHist(img)   #equalizeHist automatically tunes the image
```

```python
cv2_imshow(equ)
```

```python
plt.hist(equ.ravel(),bins = 256,range = [0,256])
plt.xlabel("Pixel Values",size = 16)
plt.ylabel("Number of Pixels",size = 16)
plt.show()
```

```python
res = np.hstack((img,equ))
```

```python
cv2_imshow(res)
```

## **Adaptive Histogram Equalization**

```python
img = cv2.imread('/content/statue.jpg',0)
```

```python
plt.hist(img.ravel(),bins = 256,range = [0,256])
plt.xlabel("Pixel Values",size = 16)
plt.ylabel("Number of Pixels",size = 16)
plt.show()
```

```python
cv2_imshow(img)
```

```python
equ = cv2.equalizeHist(img)
```

```python
plt.hist(equ.ravel(),bins = 256,range = [0,256])
plt.show()
```

```python
cv2_imshow(equ)
```

**We can clearly see that for some images, the global histogram equalizer operation is not suitable, so Adaptive Histogram Equalizer is used, actually it is doing a local operation.**

```python
clahe = cv2.createCLAHE(clipLimit = 2.0,tileGridSize = (7,7))
cl = clahe.apply(img)
```

```python
plt.hist(cl.ravel(),bins = 256,range = [0,256])
plt.xlabel("Pixel Values",size = 16)
plt.ylabel("Number of Pixels",size = 16)
plt.show()
```

```python
cv2_imshow(cl)
```

```python
final_img = np.hstack((img,equ,cl))
```

```python
cv2_imshow(final_img)
```
