```python
from PIL import Image,ImageFilter
```

```python
image = Image.open('/content/3683.jpg')
image = image.resize((300,300))
image
```

```python
blur_image = image.filter(ImageFilter.GaussianBlur(3)) #By default the size of gaussainfilter is 5 
                                                  #however we can specifiy the size within the parameters
```

```python
blur_image
```

### **We can also blur only specific part of the image**

```python
cropped_image = image.crop((55,50,240,220))
```

```python
cropped_image
```

```python
crp_blur = cropped_image.filter(ImageFilter.GaussianBlur)
```

```python
image.paste(crp_blur,(55,50,240,220))
```
